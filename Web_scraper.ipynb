## Web scraping with Beautifulsoup & Requests

Scraping through near 30,000 URL addresses, this file was copied and run in parallel across segments of the full URL lists.

key fields needing to be changed for each instance:
- start_counter
- when to end the for loop




```python
import pandas as pd
import requests
import bs4

# to get around error with some html tags being too large for the default settings.
import sys
sys.setrecursionlimit(10000)
```

```python
# this csv contains the list of webaddresses to work through.

dns = pd.read_csv('../pixc_capstone/datasets/dns_df_170926.csv')
```

```python
dns['website'] = dns.website.replace("https://","", regex=True)
dns['website'] = dns.website.replace("http://","", regex=True)
# remove if the https appears, because we're adding it in the below.
```

```python
''' This for loop takes a range of URL addresses to scrape with requests & beautifulsoup.
A csv file is automatically saved after every 200 completed rows'''

start_counter       = 10001
number_to_scrape    = 200
link_counter        = start_counter
link_counter_range  = []
all_soups           = []
all_sites           = []
scrape_status       = []


for i in range(start_counter, start_counter + number_to_scrape):
    if dns.expired_domain.iloc[i] == False:
        try:
            r = requests.get("http://" + str(dns.website.iloc[i]))
            if r.status_code != 200:
                raise ValueError
                
            soup = bs4.BeautifulSoup(r.text,'lxml')
            all_soups.append(soup) 
            print link_counter, "yes", str(dns.website.iloc[i])
            scrape_status.append("success")
            
        except:
            all_soups.append("")
            print link_counter, "no", str(dns.website.iloc[i])
            scrape_status.append("fail")
    else:
        all_soups.append("")
        print link_counter, "invalid_domain", str(dns.website.iloc[i])
        scrape_status.append("invalid_domain")
           
    all_sites.append(dns.website.iloc[i])
    link_counter_range.append(link_counter)
    
    if link_counter % 200 == 0:
        soup_df = pd.DataFrame({
            'index_range': link_counter_range,
            'site_text' : all_soups,
            'site_url' : all_sites,
            'scrape_status' : scrape_status
                })
        soup_df.to_csv('set_' + str(start_counter)+ '_webscrape_' + str(link_counter) + '.csv', encoding='utf8')
        print "\n ----------CSV_SAVED-----set_"+str(start_counter)+"_webscrape_"+str(link_counter)+".csv----------CSV_SAVED---------- \n"
    
    else:
        pass
    
    link_counter += 1
    
        
```
